# PYTHON VARIABLES:

```python
# Creating a variable:
a = 8
b = "Giri"

print(a)
print(b)
```

```python
a = 8       # a is type int now
a = "Giri"  # a is type str now

print(a)
print(type(a))
```

```python
# Casting:
a = str(42)
b = int(42)
c = float(42)

print(a)
print(b)
print(c)
```

```python
# Case-sensitive:
d = 42
D = "Boss"

print(d)
print(D)
```

### PYTHON VARIABLE NAMES:

```python
#Legal variable names:

myname = "Giri"
my_name = "Giri"
_my_name = "Giri"
myName = "Giri"
MYNAME = "Giri"
myname1 = "Giri"

print(myname)
print(my_name)
print(_my_name)
print(myName)
print(MYNAME)
print(myname1)
```

```python
#Illegal variable names:
2myname = "Giri"
my-name = "Giri"
my name = "Giri"

#This example will produce an error in the result
```

### ASSIGN MULTIPLE VALUES

```python
# Many values to multiple variables:
a, b, c = "I", "am", "developer"
print(a)
print(b)
print(c)
```

```python
#One Value to Multiple Variables:
d = e = f = "Giri The Developer"
print(d)
print(e)
print(f)
```

```python
# Unpack a collection:
mycourse = ["Python", "sql", "numpy", "pandas"]

a, b, c, d = mycourse

print(a)
print(b)
print(c)
print(d)
```

### PYTHON OUTPUT VARIABLES:

```python
g = "I am an ML Engineer"
print(g)
```

```python
g = "I"
h = "am"
i = "an"
j = "ML"
k = "Engineer"
print(g,h,i,j,k)
```

```python
# By + operator:
g = "I "
h = "am "
i = "an "
j = "ML "
k = "Engineer"
print(g + h + i + j + k)

#give space at the end of every values except lastone,
# otherwise output = IamanMLEngineer
```

```python
# + for numbers:

l = 42
m = 42

print(l + m)
```

```python
# when u combine a string and a number with + operator you got an error:

n = 3
o = "Boss"
print(n + o)
```

```python
# Seperate them with comma(,):

n = 3
o = "Boss"
print(n, o)
```

#### Other examples:

```python
name = "Giri"
age = 25
salary = 100000

print(name)
print(age)
print(salary)
```

```python
city = "Bangalore"
pincode = 560001
temperature = 28.5
is_raining = False

print(city)
print(pincode)
print(temperature)
print(is_raining)
```

```python
# Reassigning variables:
count = 10
print(count)

count = 20
print(count)

count = count + 5
print(count)
```

```python
a, b, c = 10, 20, 30
print(a)
print(b)
print(c)
```

```python
x = y = z = 100
print(x)
print(y)
print(z)
```

```python
# Variables with Arithmetic Operations:

price = 50
quantity = 4

total = price * quantity
print(total)
```

```python
#String + Variable (Concatenation):
name = "Giri"
role = "Python Developer"

print(name + " is a " + role)
```

```python
# Using Variables in f-string:
name = "Giri"
salary = 100000

print(f"My name is {name} and my salary is {salary}")
```

```python
# Swapping Variables:
a = 5
b = 10

a, b = b, a

print(a)
print(b)
```

```python
# Variable Type Checking:
value = 10.5

print(type(value))
```
