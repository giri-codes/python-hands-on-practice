# PYTHON BOOLEANS:

```python
# Ex - 1:
a = True
b = False

print(a)
print(b)
```

```python
# Ex - 2 Comparison operators:

print(10 > 5)
print(10 < 5)
print(10 == 10)
print(10 != 5)
```

```python
# Ex - 3 Boolean with Variables:

x = 20
y = 15

print(x > y)
print(x == y)
```

```python
# Ex - 4 Boolean in if Condition:

age = 18

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible")
```

```python
# Ex - 5 Logical Operators (and, or, not):
# and
print(10 > 5 and 5 > 2)
print(10 > 5 and 5 < 2)
```

```python
# Ex - 6 or:

print(10 > 5 or 5 < 2)
print(10 < 5 or 5 < 2)
```

```python
# Ex - 7 not:

print(not True)
print(not (10 > 5))
```

```python
# Ex - 8 Boolean from Functions (bool()):

print(bool(1))
print(bool(0))
print(bool("Hello"))
print(bool(""))
```

```python
# Ex - 9 Boolean with Lists:

print(bool([1, 2, 3]))
print(bool([]))
```

```python
# Ex - 10 Boolean with is Keyword:

x = True
print(x is True)
print(x is False)
```

```python
# Ex - 11:

has_id = True
has_ticket = False

print(has_id and has_ticket)
```

```python
# Ex - 12:

print(True == 1)
print(False == 0)
```

```python
# Ex - 13:

print(True + True)
print(False + True)
```
