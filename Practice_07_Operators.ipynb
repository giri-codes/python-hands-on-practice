**Time problems:**

`Full units      → //`

`Remaining units → %`

`Exact value     → /`


```python
minutes = 350

hours = minutes // 60      # full hours
remaining = minutes % 60  # remaining minutes

print(hours, remaining)
```

### OPERATOR PRECEDENCE:

`()`

`**`

`*, /, //, %`

`+, -`

`< > <= >=`

`== !=`

`not`

`and`

`or`

```python
# Parentheses():
a = 5
b = 3
result = (a + b) * 2
print(result)

# Execution order:
# (5 + 3) → 8 → 8 * 2 = 16
```

```python
# Power **

a = 2
b = 3
result = a ** b + 1
print(result)

# Order:
# 2 ** 3 = 8 → 8 + 1 = 9
```

```python
# Multiplication before Addition * then +

a = 10
b = 5
c = 2

result = a + b * c
print(result)

# Order:
# 5 * 2 = 10 → 10 + 10 = 20
```

```python
# Division vs Floor Division / vs //

total_minutes = 350
hours_exact = total_minutes / 60
hours_full = total_minutes // 60

print(hours_exact)
print(hours_full)
```

```python
# Comparison before Logical > < == before and or

a = 10
b = 5
c = 3

result = a > b and b > c
print(result)

# Order:
# 10 > 5 → True
# 5 > 3 → True
# True and True → True
```

```python
# not before and:
is_raining = False
is_sunny = True

result = not is_raining and is_sunny
print(result)

# Order:
# not False → True
# True and True → True
```

```python
# and before or:

a = False
b = True
c = False

result = a or b and c
print(result)

# Order:
# b and c → False
# a or False → False
```

```python
# Mixed precedence:

x = 10
y = 2
z = 3

result = x + y * z ** 2
print(result)

# Order:
# z ** 2 → 9
# y * 9 → 18
# x + 18 → 28
```

```python
# Using parentheses to change precedence:

x = 10
y = 2
z = 3

result = (x + y) * z ** 2
print(result)

# Order
# (10 + 2) → 12
# 3 ** 2 → 9
# 12 * 9 → 108
```

```python
# Comparison + Logical + not:

age = 25
experience = 1

eligible = not age < 21 and experience >= 1
print(eligible)

# Order:
# age < 21 → False
# not False → True
# experience >= 1 → True
# True and True → True
```
