### LOOP THROUGH DICTIONARIES:

```python
# loop keys:
book = {
    "title": "Python",
    "author": "XYZ",
    "price": 500
}

for key in book:
    print(key)
```

```python
# Loop keys by key():
marks = {
    "Maths": 80,
    "Science": 85
}

for e in marks.keys():
    print(e)
```

```python
# Loop values:
book = {
    "title": "Python",
    "author": "XYZ",
    "price": 500
}

for value in book:
    print(book[value])
```

```python
# Loop values by value():
marks = {
    "Maths": 80,
    "Science": 85
}

for value in marks.values():
    print(value)
```

```python
# Loop keys and values by items():

employee = {
    "name": "Kiran",
    "salary": 30000,
    "dept": "IT"
}

for key, value in employee.items():
    print(key, ":", value)
```

### NESTED DICTIONARIES:

```python
# Bank accounts:
bank = {
    "acc1": {"name": "Raju", "balance": 5000},
    "acc2": {"name": "Meena", "balance": 10000}
}

print(bank["acc2"]["balance"])
print(bank)
```

```python
# College students:
college = {
    "student1": {"name": "Asha", "marks": 90},
    "student2": {"name": "Ravi", "marks": 85},
    "student3": {"name": "Meena", "marks": 88}
}

print(college["student2"]["name"])
print(college["student3"]["marks"])
```

```python
# Loop nested dictionary - outer keys:
employees = {
    "E1": {"name": "Raj", "salary": 30000},
    "E2": {"name": "Anu", "salary": 45000}
}

for emp_id in employees:
    print(emp_id)
```

```python
# Loop nested dictionary - outer key + inner dictionary:
# print full employee data:

employees = {
    "E1": {"name": "Raj", "salary": 30000},
    "E2": {"name": "Anu", "salary": 45000}
}

for emp_id, details in employees.items():
    print(emp_id, details)
```

```python
# Loop inner dictionary (key + value):
# Print Each Employee Details Clearly:

e = {
    "E1": {"name": "Giri", "salary": 100000},
    "E2": {"name": "Rakshitha", "salary": 45000}
}

for e_id, details in e.items():
    print("Employee:", e_id)
    for key, value in details.items():
        print(key, ":", value)
```

```python
# Access using loop + condition:
# print salary only:
employees = {
    "E1": {"name": "Raj", "salary": 30000},
    "E2": {"name": "Anu", "salary": 45000}
}

for details in employees.values():
    print(details["salary"])
```

```python
# Nested dictionary with subjects:
# student marks:
students = {
    "Giri": {"Maths": 90, "Science": 85},
    "Rekha": {"Maths": 88, "Science": 92}
}

for name, subjects in students.items():
    print(name)
    for subject, marks in subjects.items():
        print(subject, "=", marks)
```

```python
# 3-Level nested dictionary:
# School → Class → Student:

s = {
    "Class10" : {
        "student1" : {"name" : "Giri", "marks" : 99},
        "student2" : {"name" : "Rakshitha", "marks" : 98}
    }
}

print(s["Class10"]["student2"]["name"])
```

```python
# Full company data:

c = {
    "Company" :{
        "employee1" : {"name" : "giri", "salary" : 100000},
        "employee2" : {"name" : "basava", "salary" : 50000}
    }
}

for company_name, employees in c.items():
    print(company_name)
    for employee_id, details in employees.items():
        print(employee_id)
        for key, value in details.items():
            print(key, ":", value)
```

```python
# Bank data loop:
bank = {
    "acc1": {"name": "Raju", "balance": 5000},
    "acc2": {"name": "Meena", "balance": 12000}
}

for acc_no, info in bank.items():
    print("Account:", acc_no)
    print("Name:", info["name"])
    print("Balance:", info["balance"])
```
