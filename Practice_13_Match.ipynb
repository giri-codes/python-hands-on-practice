`_ means else / default`

```python
a = 8

match a:
    case 1:
        print("odd")
    case 2:
        print("even")
    case 3:
        print("odd")
    case 4:
        print("even")
    case 5:
        print("odd")
    case 6:
        print("even")
    case 7:
        print("odd")
    case 8:
        print("even")
    case _:
        print("Not a number")
```

```python
# Ex - 2 menu selection:

choice = 3

match choice:
    case 1:
        print("You ordered idli")
    case 2:
        print("You ordered dosa")
    case 3:
        print("You ordered poori")
    case _:
        print("Invalid choice")
```

```python
# match vs if - elif:
#using if - elif:
if choice == 1:
    print("Idli")
elif choice == 2:
    print("Dosa")
elif choice == 3:
    print("Poori")
else:
    print("Invalid")
```

```python
# Using match:

match choice:
    case 1:
        print("idli")
    case 2:
        print("dosa")
    case 3:
        print("poori")
    case _:
        print("Invalid")
```

```python
# Multiple values in once case:

day = "Saturday"

match day:
    case "Saturday" | "Sunday":
        print("Weekend")
    case _:
        print("working day")
```

```python
# Vowels check:

ch = "e"

match ch:
    case "a" | "e" | "i" | "o" | "u":
        print("vowels")
    case _:
        print("consonant")
```

```python
# Using varibales inside case:
num = 42

match num:
    case 0:
        print("zero")
    case a if a > 0:
        print("Positive number")
    case _:
        print("Negative number")
```

```python
#Match with strings:

role = "admin"

match role:
    case "admin":
        print("full access")
    case "user":
        print("Limited access")
    case "guest":
        print("read-only access")
    case _:
        print("unknown role")
```

```python
# Number type check:

n1 = 4

match n1:
    case n if n > 0:
        print("positive number")
    case n if n < 0:
        print("negative number")
    case _:
        print("zero")
```

```python
age = 27

match age:
    case m if m < 18:
        print("minor")
    case m if m >= 18 and m < 60:
        print("adult")
    case _:
        print("senior")
```

```python
# Login validation:

user = "giri"
password = "4123"

match user:
    case "giri" if password == "4123":
        print("login success")
    case "admin":
        print("wrong password")
    case _:
        print("invalid user")
```

```python
day = "sunday"

match day:
    case "saturday" | "sunday" if day == "sunday":
        print("relax fully")
    case "saturday" | "sunday":
        print("weekend vibes")
    case _:
        print("work day")
```

```python
month = 5
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5 if month == 4:
    print("A weekday in April")
  case 1 | 2 | 3 | 4 | 5 if month == 5:
    print("A weekday in May")
  case _:
    print("No match")
```

```python
m = 4
d = 8

match d:
    case 1|2|3|4|5|6|7|8 if m == 4:
        print("Boss birthday")
    case 1|2|3|4|5 if m == 7:
        print("sister birthday")
    case _:
        print("no match")
```
