# PYTHON WHILE LOOP:

**SYNTAX**:

`while condition:`

    statements


```python
# Print numbers 1 to 5:
i = 0
while i <= 9:
    print(i)
    i += 1
```

```python
# Print "giri" 3 times:

count = 1

while count <= 3:
    print("giri")
    count += 1
```

```python
# Print even numbers till 10:

num = 2

while num <= 10:
    print(num)
    num += 2
```

```python
# Print odd numbers till 10:

num1 = 1

while num1 <= 10:
    print(num1)
    num1 += 2
```

```python
# Using break:
i = 1

while i <= 10:
    if i == 8:
        break   #break = stop loop immediately
    print(i)
    i += 1
```

```python
# Using continue:
i = 0

while i < 5:
    i += 1
    if i == 3:
        continue  # continue = skip current iteration
    print(i)
```

```python
# Else with while loop:
i = 1
while i <= 4:
    print(i)
    i += 1
else:          # else runs only if loop ends normally (no break)
    print("Loop completed")
```

```python
# Print multiplication table of 8:
i = 1

while i <= 10:
    print("8 x", i, "=", 8 * i)
    i += 1
```

```python
# Countdown from 10 to 1:
i = 10

while i >= 1:
    print(i)
    i -= 1
```

```python
# top loop when number reaches 7 (using break):

i = 1

while i <= 25:
    if i == 7:
        break
    print(i)
    i += 1
```

```python
# Skip number 5 (using continue):
i = 0

while i < 10:
    i += 1
    if i == 5:
        continue
    print(i)
```

```python
i = 1
total = 0

while i <= 5:
    total += i
    i += 1

print("Sum =", total)
```

```python
# Print digits of a number*:

num = 1234

while num > 0:
    digit = num % 10
    print(digit)
    num //= 10
```

```python
# Password retry:
password = "python"
attempt = " "

while attempt != password:
    attempt = input("Enter password: ")

print("Access granted")
```
