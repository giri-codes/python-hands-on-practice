### DEFAULT ARGUMENTS:

`Default argument = backup value`

`Used only when no argument is passed`


```python
# ex - 1: simple default:
def welcome(name = "Friend"):
    print("welcome", name)

welcome()
welcome("Giri")
```

```python
# ex - 2 default number:
def power(base, exp = 2):
    print(base ** exp)

power(3)
power(3,4)
```

```python
# ex - 3 multiple defaults:
def profile(name = "user", city = "India"):
    print(name)
    print(city)

profile()
profile("Giri")
profile("Giri", "Mysuru")
```

```python
# ex - 4 default in condition:
def login(user = "admin"):
    if user == "admin":
        print("Admin access")
    else:
        print("User access")

login()
login("giri")
```

```python
# ex - 5 mixing normal + default:
def greet(name, msg="Good morning!"):
    print(msg, name)

greet("Giri")
greet("Giri", "Hello")
```

### *args:

```python
# ex -1:
def show(*args):
    print(args)

show(1)
show(1,2)
show(1,2,3,3,4)
```

```python
# ex - 2 loop through args:
def print_values(*args):
    for value in args:
        print(value)

print_values("giri", "python", "ML")
```

```python
# ex -3 sum numbers:
def total(*nums):
    s = 0
    for n in nums:
        s += n
    print(s)

total(10, 20)
total(1,2,3,4,5)
```

```python
# ex - 4 mixed noraml + *args:
def greet(name, *msgs):
    print("Name:", name)
    print("Messages:", msgs)

greet("Giri", "Hi", "Welcome", "Good luck")
```

```python
# ex - 5
def demo(*args):
    print(args)

demo(100) # args is a tuple
```

### **kwargs:

```python
# ex -1 print all key-value pairs:
def show_details(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

show_details(name = "Giri", role = "Developer", skill = "Python")
```

```python
# ex - 2 access specific value:
def show_name(**info):
    print(info["name"])

show_name(name = "Giri", age = 25)
```

```python
# ex - 3 mixed normal + **kwargs:
def student(name, **details):
    print("Name:", name)
    print("Other details:", details)

student("Giri", age = 25, city = "Mysuru", course = "Python")
```

```python
# ex - 4 conditional use:
def login(**data):
    if data.get("user") == "admin":
        print("Admin login")
    else:
        print("user login")

login(user = "admin")
login(user = "giri")
```

```python
# ex - 5 empty kwargs allowed:
def demo(**kwargs):
    print(kwargs)

demo()
```

### PYTHON SCOPE:

```python
# ex - 1 local variable:
def test():
    a = 5
    print(a)

test()
```

```python
# ex - 2 global variable:
b = 10
def test():
    print(b)

test()
```

```python
# ex - 3 local vs global:
x = 100

def demo():
    x = 50 # local x
    print(x)

demo()
print(x)
```

```python
# ex - 4 same name different scope:

x = 10   # global variable

def test():
    x = 5    # local variable
    print("Inside function:", x)

test()
print("Outside function:", x)
```

```python
# ex - 5 reading vs modifying global:
count = 0   # global variable

def show():
    print(count)   # reading global is allowed

def increase():
    global count   # tells Python to use global variable
    count = count + 1

show()
increase()
show()

```

### LAMBDA FUNCTIONS:

### DECORATORS:

```python
# ex-1:
def drink():
    print("I am drinking milk")

def g(func): 
    def wrapper():
        print("I Added sugar")
        func()
        print("I Clean the glass")
    return wrapper

@g
def drink():
    print("I am drinking milk")

drink()
```

```python
# ex-2:
def the_decorator(func):
    def wrapper():
        print("Start")
        func()
        print("End")
    return wrapper

@the_decorator
def hello():
    print("Hello")

hello()
```

```python
# ex- 3 decorator for security:
def check_login(func):
    def wrapper():
        print("Checking login...")
        func()
    return wrapper

@check_login
def dashboard():
    print("Dashboard opened")

dashboard()
```

```python
# ex- 4 decorator without @ (manual):
def log(func):
    def wrapper():
        print("Function called")
        func()
    return wrapper

def task():
    print("Task running")

task = log(task)
task()
```

```python
# ex - 5:
def my_decorator(func):
    def wrapper():
        print("Start")
        func()
        print("End")
    return wrapper

@my_decorator
def task():
    print("Task loading")

task()
```

### PYTHON RECURSION:

```python
# ex- 1:
def show(n):
    if n == 0:
        return
    print(n)
    show(n - 1)

show(10)
```

```python
# ex - 2 sum of numbers:
def sum_1(n):
    if n == 0:
        return 0
    return n + sum_1(n - 1)
print(sum_1(3))
```

```python
# ex - 3 factorial:
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
# ex - 4:
def add(n):
    if n == 0:
        return 0
    return n + add(n-1)
add(10)
```

### GENERATORS:

```python
# ex - 1 Generator for countdown timer:
def countdown(seconds):
    while seconds > 0:
        yield seconds
        seconds -= 1

for t in countdown(6):
    print("Time left:", t)
```

```python
# ex - 2 Generator for reading tasks:
def tasks():
    yield "Wake up"
    yield "Study Python"
    yield "Exercise"
    yield "Sleep"

for task in tasks():
    print(task)
```

```python
# ex - 3 Generator for squares of numbers:
def squares(n):
    for i in range (1, n+1):
        yield i * i

for s in squares(5):
    print(s)
```

```python
# ex - 4 Generator for password attempts:
def attempts():
    yield "1234"
    yield "admin"
    yield "password"
    yield "giri@123"

for pd in attempts():
    print("Typing:", pd)
```

```python
# ex - 5 Generator for odd numbers:
def odd_numbers(lm):
    for i in range(1, lm + 1, 2):
        yield i

for o in odd_numbers(7):
    print(o)
```
