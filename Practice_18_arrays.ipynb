# PYTHON ARRAYS:
In Python, an array is implemented using a list, which is an ordered, mutable collection used to store multiple elements in a single variable.

```python
# Example 1: Daily expenses tracking
expenses = [120, 250, 80, 60]
total = sum(expenses)
print(total)
```

```python
# Example 2: Student marks storage
marks = [85, 90, 78, 88]
highest = max(marks)
print(highest)
```

```python
# Example 3: Shopping cart items
cart = ["milk", "bread", "eggs"]
cart.append("butter")
print(cart)
```

```python
# Example 4: Attendance status
attendance = [True, True, False, True]
absent_count = attendance.count(False)
print(absent_count)
```

```python
# Example 5: City visit list
cities = ["Bangalore", "Mysuru", "Chennai"]
if "Mysuru" in cities:
    print("City found")
```
